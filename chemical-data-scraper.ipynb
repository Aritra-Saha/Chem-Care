```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pprint import pprint
from difflib import SequenceMatcher
import json
```

```python
page_dir = 'https://www.ewg.org'
num_pages = 756 # 756
```

```python
links = []
lol_count = 0
for n in range(1, num_pages+1):
    link = "https://www.ewg.org/skindeep/search?page=%d&search=&search_type=ingredients" % n
    c = 0
    while c == 0:
        response = requests.get(link, timeout=5)
        soup = BeautifulSoup(response.content, "html.parser")
        tags = soup.findAll("div", attrs={"class": "product-tile"})
        c = len(tags)
    for tag in tags:
        next_link = tag.a["href"]
        if next_link != "":
            links.append(next_link)
        else:
            lol_count += 1
    print(n,c)
```

```python
chemical_data = {}
id_map = {}
id_n = 0
```

```python
def add_data(link):
    global id_n
    id_n += 1
    print(link)
    response = requests.get(link, timeout=5)
    soup = BeautifulSoup(response.content, "html.parser")
    #print(soup)
    while True:
        try:
            synonyms = soup.find("p", attrs={"class":"chemical-synonyms-text"}).getText().split(", ")
            break
        except:
            print("error")
            response = requests.get(link, timeout=5)
            soup = BeautifulSoup(response.content, "html.parser")
        
    synonyms = synonyms[:-1] + [synonyms[-1][4:]]
    print(synonyms)
    concerns = {}
    references = {}
    for div in soup.findAll("section", attrs={"class": "collapsable-block"}):
        title = div.h2.getText()
        if title != "Data gaps" and title != "Products with this Ingredient":
            i = 0
            concerns[title] = []
            references[title] = []
            for x in div.table.findAll("td"):
                if (i % 2) == 0:
                    concerns[title].append(x.getText())
                else:
                    references[title].append(x.getText())
                i += 1
            print(concerns[title])
            print(references[title])
    for syn in synonyms:
        id_map[syn] = id_n
        
    chemical_data[id_n] = [concerns, references]
```
